```python
import numpy as np
import numpy.random as random
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
sns.set()
%matplotlib inline

%precision 3
```

```python
#棒グラフ(bar関数)

x = [1,2,3]
y = [10,1,4]

plt.figure(figsize = (10,6))

plt.bar(x,y,align = 'center',width = 0.5)

#棒グラフそれぞれのラベル
plt.xticks(x,['A','B','C'])

#x,yのラベル
plt.xlabel('Class')
plt.ylabel('Score')

plt.grid(True)
```

```python
#横の棒グラフ(barh関数)
x = [1,2,3]
y = [10,1,4]

plt.figure(figsize = (10,6))

plt.barh(x,y, align = 'center')
#横の棒グラフのため，ラベルを入れ替える
plt.yticks(x,['A','B','C'])
plt.ylabel('Class')
plt.xlabel('Score')

plt.grid(True)
```

```python
#複数の棒グラフを描画

y1 = np.array([30,10,40])
y2 = np.array([10,50,90])

x = np.arange(len(y1))

#グラフの幅
w = 0.4

plt.figure(figsize =(10,6))

plt.bar(x, y1,color ='blue', width = w, label = 'Math first', align ='center')
#y2はグラフの幅の分右にずらす
plt.bar(x + w,y2,color ='green', width = w, label ='Math final', align = 'center')

#凡例を最適な位置へ
plt.legend(loc = 'best')
plt.xticks(x + w / 2,['A','B','C'])
plt.grid(True)
```

```python
#積み上げ棒グラフ(barのパラメータでbottomを指定)
height1 = np.array([100,200,300,400,500])
height2 = np.array([1000,800,600,400,200])

x = np.arange(len(height1))

plt.figure(figsize = (10,6))

p1 = plt.bar(x, height1, color = 'blue')
p2 = plt.bar(x,height2, bottom = height1,color = 'lightblue')

plt.legend((p1[0],p2[0]),('Class 1','Class 2'))
```

```python
#円グラフ(pie関数)
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15,30,45,10]
colors =['yellowgreen','gold','lightskyblue','lightcoral']
explode = (0,0.1,0,0)

plt.figure(figsize = (15,6))

#autopctパラメータで割合表示の書式を指定
#explodeパラメータで特定のカテゴリーを円から離せる
#startangleパラメータは各要素の出力開始位置を変更できる
plt.pie(sizes, explode = explode, labels = labels, colors = colors,
        autopct ='%1.1f%%', shadow = True, startangle = 90,counterclock = False)

#円を丸く書く描画(axis関数)
plt.axis('equal')
```

```python
#バブルチャート(scatter関数)
N = 25

#random生成
x = np.random.rand(N)
y = np.random.rand(N)

colors = np.random.rand(N)

#バブルの大きさをばらけさせる
area = 10 * np.pi * (15 * np.random.rand(N)) **2

plt.figure(figsize =(15,6))

plt.scatter(x,y, s = area, c = colors, alpha = 0.5)
plt.grid(True)
```

```python
#dateをdate型に変更
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
#.str は Series（1列） にしか使えません
#apply を使って 各列に対して処理
#.str.replace('$','', regex=False) で
# regex=False は str.replace() において、置換対象を「文字列そのもの」として扱うためのオプションです。
#.astype(float) で文字列 → 数値に変換

cols = ['open','high','low','close','next_weeks_open','next_weeks_close']
df[cols] = df[cols].apply(lambda x : x.astype(str).str.replace('$','',regex = False).astype(float))
df
```
