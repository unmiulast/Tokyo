```python
data1 = {
    'id':['100','101','102','103','104','106','108','110','111','113'],
    'city':['Tokyo','Osaka','Kyoto','Hokkaido','Tokyo','Tokyo','Osaka','Kyoto','Hokkaido','Tokyo'],
    'birth_year':[1990,1989,1992,1997,1982,1991,1988,1990,1995,1981],
    'name':['Hiroshi','Akiko','Yuki','Satoru','Steeve','Mituru','Aoi','Tarou','Suguru','Mitsuo']
}

df1 = DataFrame(data1)
df1
```

```python
data2 = {
    'id':['100', '101', '102', '105', '107'],
    'math':[50,43,33,76,98],
    'English':[90,30,20,50,30],
    'sex':['M','F','F','M','M'],
    'index_num':[0,1,2,3,4]
}

df2 = DataFrame(data2)
df2
```

```python
#内部結合(mergeのデフォルト,キーは自動認識,onで明示可能)
pd.merge(df1,df2, on = 'id')
```

```python
#全結合(howパラメータにouterを指定)
pd.merge(df1,df2,how = 'outer')
```

```python
#インデックスによるmerge
#これは「df1 のインデックス」と「df2['index_num'] の列」を結合キーにしています。
pd.merge(df1,df2, left_index = True, right_on = 'index_num')
```

```python
#左外部結合(howパラメータにleftを指定)
pd.merge(df1,df2,how = 'left')
```

```python
#時系列データ
start_date = '2001/1/2'
end_date ='2016/12/30'

fx_jpusdata = pdr.DataReader('DEXJPUS','fred',start_date, end_date)
```

```python
fx_jpusdata
```

```python
#特定の年月を参照
fx_jpusdata.loc['2016-04']
```

```python
#'MS' → 月初 (MonthStart)
#'ME' → 月末 (MonthEnd) ← 今後はこちらを推奨
fx_jpusdata.resample('ME').last().head()
```

```python
#日付を取り出したい時は'D',年を取り出したい時は'Y'
fx_jpusdata.resample('D').last().head()
```

```python
#欠損を前日で埋める
fx_jpusdata.resample('D').ffill().head()
```

```python
#データをずらして比率を計算する
#shiftメソッドを使うことでindexは固定したままデータだけをずらせる

fx_jpusdata.shift(1).head()
```

```python
#前日と当日のレートの比を算出
fx_jpusdata_ratio = fx_jpusdata / fx_jpusdata.shift(1)
fx_jpusdata_ratio.head()
```

```python
fx_jpusdata.resample('YE').mean()
```

```python
#移動平均
fx_jpusdata.head()
```

```python
#rollingメソッドを実行後，meanメソッド
fx_jpusdata.rolling(3).mean()
```

```python
#標準偏差の推移はstdメソッド
fx_jpusdata.rolling(3).std().head()
```
